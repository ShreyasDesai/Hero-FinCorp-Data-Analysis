```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
branches = pd.read_csv("../data/branches_cleaned.csv")
loans = pd.read_csv("../data/loans_cleaned.csv")
```

```python
region_active_loans = branches.groupby('Region')['Total_Active_Loans'].sum()
region_active_loans
```

```python
region_active_loans.plot(kind='bar', figsize=(7,4))
plt.title("Active Loans Distribution by Region")
plt.ylabel("Total Active Loans")
plt.xticks(rotation=45)
plt.show()
```

```python
region_default = branches.groupby('Region').agg({
    'Delinquent_Loans': 'sum',
    'Total_Active_Loans': 'sum'
})

region_default['Default_Rate'] = (
    region_default['Delinquent_Loans'] /
    region_default['Total_Active_Loans']
)

region_default
```

```python
region_default['Default_Rate'].plot(kind='bar', figsize=(7,4))
plt.title("Default Rate by Region")
plt.ylabel("Default Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
region_disbursement = branches.groupby('Region')['Loan_Disbursement_Amount'].sum()
region_disbursement
```

```python
region_disbursement.plot(kind='bar', figsize=(7,4))
plt.title("Loan Disbursement Volume by Region")
plt.ylabel("Disbursement Amount")
plt.xticks(rotation=45)
plt.show()
```

### Task 13 – Geospatial Insights

- Loan activity varies significantly across regions
- Certain regions dominate in active loan volume and disbursement
- Default rates show geographic concentration, indicating regional risk patterns
- Region-specific credit policies can help reduce default exposure
